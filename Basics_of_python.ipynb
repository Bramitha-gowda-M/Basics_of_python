Boolean Variables

Boolean values are the two constant objects FALSE and TRUE.

They are used to represent truth values(other values can also be considered false or true.)

In numeric contexts(for example, when used as the argument to an arithmetic operator), they behave like the integers 0 and 1, respectively.

The built-in functions bool() can be used to cast any value to a Boolean, If the value can be interpreted as a truth value.

They are written as False and True, respectively.

```python
print(True,False)
```

```python
type(True)
```

```python
type(False)
```

```python
my_str = "Bramitha"

print(my_str.isalpha())
print(my_str.isalnum())
print(my_str.islower())
print(my_str.isupper())
print(my_str.isspace())
print(my_str.istitle())
print(my_str.endswith("a"))
print(my_str.startswith("B"))
print(my_str.isdigit())
```

```python
mystr = "BramithaGowd"
print(mystr.isalpha())
```

# Sets

A set is an unordered collection data type that is iterable, mutable and has no duplicate elements. Pythons set class represents the mathematical notion of a set. This is based on a data structure known as a has table

```python
# Defining an empty set

set_var = set()
print(set_var)
print(type(set_var))
```

```python
set_var = {"Avengers","IronMan","HitMan"}
print(set_var)
print(type(set_var))
```

```python
#sets doesn't support the indexing retrieving
```

```python
## inbuilt function in sets

set_var.add("Hulk")
```

```python
print(set_var)
```

```python
set1 = {1,2,3}
set2 = {3,4,5}
```

```python
set2.difference(set1)
```

```python
## difference update

set2.difference_update(set1)
```

```python
print(set2)
```

#Dictionaries

A dictionary is a collection which is unordered, changeable and indexed. In python dictionaries are written with curly brackets, and they have keys and values.

```python
dic = {}
```

```python
type(dic)
```

```python
dic={1,2,3,4,5}
```

```python
type(dic)
```

```python
## lets create a dictionary

my_dict =  {"car1":"Audi","car2":"BMW","car3":"Mercedez"}
```

```python
type(my_dict)
```

```python
## Access the item values based on keys

my_dict['car1']
```

```python
#we can even loop throught the dictionaries keys

for x in my_dict:
  print(x)
```

```python
#we can even loop throught the dictionaries keys

for x in my_dict.values():
  print(x)
```

```python
#we can even loop throught the dictionaries values and keys

for x in my_dict.items():
  print(x)
```

```python
#adding items in Dictionaries

my_dict['car4']="Honda"
```

```python
my_dict
```

#Nested Dictionary

```python
car1_model = {"Mercedez":1960}
car2_model = {"Audi":1970}
car3_model = {"BMW":1980}

car_type = {"car1":car1_model,"car2":car2_model,"car3":car3_model}
```

```python
print(car_type)
```

```python
## Accessing the items in the dictionary

print(car_type['car1'])
```

```python
print(car_type['car1']['Mercedez'])
```
